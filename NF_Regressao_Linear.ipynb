```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.array([4.2, 4.5, 4.9, 5.5, 4.9, 4.4, 4.7, 5.2, 5.5, 4.6]) #em trilhões de R$
y = np.array([6.8, 6.9, 7.5, 7.6, 8.5, 9.8, 10.3, 11.1, 11.9, 13.5]) #em porcentagem

coeficiente_angular, intercepto = np.polyfit(x, y, 1)

y_pred = coeficiente_angular * x + intercepto

plt.scatter(x, y, label = 'Dados originais')
plt.plot(x, y_pred, color = 'black', linewidth = 2, label = 'Linha de Regressao')
plt.xlabel('PIB em trilhoes')
plt.ylabel('Desemprego em porcentagem')
plt.legend()
plt.title('Relação entre PIB e taxa de desemprego')
plt.grid(True)
plt.show()

x = pd.DataFrame({'Valores': [4.2, 4.5, 4.9, 5.5, 4.9, 4.4, 4.7, 5.2, 5.5, 4.6]})
y = pd.DataFrame({'Valores': [6.8, 6.9, 7.5, 7.6, 8.5, 9.8, 10.3, 11.1, 11.9, 13.5]})

x_media = x['Valores'].mean()
x_mediana = x['Valores'].median()
x_moda = x['Valores'].mode()
x_std = x['Valores'].std()

y_media = y['Valores'].mean()
y_mediana = y['Valores'].median()
y_moda = y['Valores'].mode()
y_std = y['Valores'].std()

print(x.describe())
print(y.describe())
```
